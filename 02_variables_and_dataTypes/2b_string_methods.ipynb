# **String Methods**
Python provides a set of built-in methods that we can use to alter and modify the strings.

## **`upper()`:**

The `upper()` method converts a string to upper case.

**Example:**

```python
str1 = "AbcDEfghIJ"
print(str1.upper())
```

**Output:**

```python
# ABCDEFGHIJ
```

## **`lower()`:**

The `lower()` method converts a string to lower case.

**Example:**

```python
str1 = "AbcDEfghIJ"
print(str1.lower())
```

**Output:**

```python
# abcdefghij
```

## **`strip()`:**

- Removes **characters from both ends** (start and end) of a string.

- By default, it removes **whitespace**, but you can specify any characters.

**Example:**

```python
s = "  Hello World  "
print (s.strip()) # remove spaces from both ends of string

# Output:
# 'Hello World'
```

**Example 2**

```python
s = "  !!!$Hello World$$!!!  "
stripped = s.strip(" !$") # will remove spaces (!,$) from both sides
print ("Original:", repr(s))
print ("Stripped:", repr(stripped))
```

**Output:**

```python

# Original: '  !!!$Hello World$$!!!  '
# Stripped: 'Hello World'
```

**Example 3:**

```python
str = "......... Sadaqat Ali "
stripForDot = str.strip(".") # it will just remove (.) not spaces.
StripForDotSpace = str.strip(". ") # it will remove both dots and spaces
print ("Original:", repr(str))
print ("Stripped for Dots:", repr (stripForDot)) 
print ("Stripped for Dots and Spaces:", repr (StripForDotSpace)) 
```

**Output:**

```python
# Original: '......... Sadaqat Ali '
# Stripped for Dots: ' Sadaqat Ali '
# Stripped for Dots and Spaces: 'Sadaqat Ali'
```

## **`rstrip()`:**

- Removes characters **only from the right end** (the end of the string).

- Default is whitespace, but you can specify characters.

**Example:**

```python
str = "Hello World!!!"
rstriped = str.rstrip("!") # will remove exclamation marks form right end of str
print ("Original:", repr (str))
print ("Rstripped:", repr (rstriped))
```

**Output:**

```python
# Original: 'Hello World!!!'
# Rstripped: 'Hello World'
```

**Example 2:**

```python
s = "  Hello World!!!    "
print(repr(s.rstrip()))       # removes spaces from the right only
print(repr(s.rstrip(" !")))   # removes spaces and ! from the right only
```

**Output:**

```python
'  Hello World!!!'
'  Hello World'
```

## **`lstrip()`:**
- Removes characters **only from the left/start** of the string.

**Example:**


```python
s = "!!!  Hello World   "
s.lstrip(" !")  # removes spaces and ! from the left only
```

**Output:**

```python
'Hello World   '
```

The `capitalize()` method turns only the first character of the string to uppercase and the rest other characters of the string are turned to lowercase. The string has no effect if the first character is already uppercase.

**Example:**

```python
str1 = "hello"
print (str1.capitalize())
str2 = "hello WorlD"
print (str2.capitalize())
```

## **`center()`:**

- Centers a string within a given width.

- Optionally fills extra space with a specified character.

**Syntax:**

```python
# string.center(width, fillchar)
```

**width** → total width of the resulting string.

**fillchar** → optional character to fill spaces (default is space ' ').

**Let’s Suppose:** 

- Total width is 11, original text is 5 characters, so 6 spaces are divided equally on both sides.




**Example:**

```python
str1 = "Hello"
print(str1.center(11)) # Total width is 11, original text is 5 characters, so 6 spaces are divided equally on both sides.
```

**Output:**

```python
#    Hello
```

We can also provide a padding character. It will fill the rest of the fill characters provided by the user.

**Example:**

```python
str1 = "Welcome to the Console!!!"
print(str1.center(50, ".")) # total width is 50, original text is 25 characters, so 25 dots are divided equally on both sides
```

**Output:**

```python
# ............Welcome to the Console!!!.............
```

## **`endswith()`:**

The `endswith()` method checks if the string ends with a given value. If yes, then return True, else return False.

**Example 1: Without start/end (whole string):**

```python
str1 = "Hello"
print(str1.endswith("o"))  # True, last character is 'o'
print(str1.endswith("H"))  # False, first character is 'H'

str = "how are you"
print (str.endswith("you"))
```

**Output:**

```python
# True
# False
# True
```

- Checks the whole string if it ends with the given suffix.

- If suffix is one character, it can return True for that single character.

**Example 2: With start and end (slice):**

```python
str = "Welcome to the Console !!!"
print(str.endswith("to",4 , 10)) 
str1 = "How are you"
print (str1.endswith("w",0 , 3)) # will creat slice [0:3] 'How' 
```

**Output:**

```python
# True
# True
```

- Python first creates the `slice: str[4:10] = 'ome to'`

- Then it checks if the **entire slice ends with the suffix**.



## **`find()`:**

### **37. Write a python program to check if a string is palindrome or not.**

```python
a = input ("Enter a string: ")
b = a[::-1]
if a == b:
    print (f"\"{a}\" is palindrome string.")
else:
    print (f"\"{a}\" is not palindrome string.")
```

```python
num = int(input("Enter a number: "))

original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print(f"{original} is Palindrome.")
else:
    print(f"{original} is not Palindrome.")
```
