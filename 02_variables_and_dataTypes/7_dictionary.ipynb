## ***Dictionary (`dict`)***

- Dictionary allows user to write the data in the form of keys and values.

- Dictionaries are enclosed inside curly brackets {}.

- Keys and Values are separated by colon.

- Every Key Value pair are separated by coma(,).

- `dict` are mutable, we can modify them like `list`.

```python
employee_data = {
    "Name" : "Jhon",
    "Age" : 24,
    "Gender" : "Male"
}

print (employee_data) # it will print whole dictionary

print (employee_data["Gender"]) # it will print just the value of Key "Gender".
```

## **Iteration in Dictionary**

**1. Printing all key names one by one.**

```python
student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x in student:
    print (x) # it will print just keys
```

**2. Printing all value names one by one.**

```python
# Solution 1

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for i in student: # it will access each key in dictionary one by one
    print (student[i]) # it will print print the values of keys stored in student variable, just like to get only "Value" of "Key" --> Name! we can type 
    # print (student["Name"])
```

```python
# Solution 2 Using value function

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x in student.values(): # it will print just values in the dictionary one by one
    print (x)
```

```python
# Using items function (printing both keys and values)

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x in student.items(): 
    print (x)
```

```python
# Using items function (Printing just keys)


student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x,y in student.items(): 
    print (x) # it will print just keys
```

```python
# Using items function (Printing just values)


student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x,y in student.items(): 
    print (y) # it will print just values
```

```python
# Using items function (Printing just values)

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x,y in student.items(): 
    print (y) # it will print just values
```

```python
# Using items function (Printing both keys and values using both)

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

for x,y in student.items(): 
    print (x,":",y) # it will print both keys and values separated by colon
```

## **Dictionary Functions**

```python
# get function

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

print (student.get("Name")) # it will print just the value of key "Name"
```

```python
# items functoin

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

print (student.items()) # it will print both keys and values separated by coma(,)
```

```python
# keys function

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

print (student.keys()) # it will print only keys of the dictionary
```

```python
# values function

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

print (student.values()) # it will print only values fo the dictionary
```

```python
# copy function

student = {
    "Name" : "Jhon",
    "Class" : "6th", 
    "Roll_No" : 23
    }

data = student.copy() # it will copy the student into data variable

print (data) # it will print copied form of student dictionary
```

```python
# setdefault function

student = {
    "Name": "Jhon",
    "Class": "6th",
    "Roll_No": 23
}
# Dictionary 'student' already contains the key "Roll_No"

a = student.setdefault("Roll_No", 24)
# setdefault() checks if the key "Roll_No" exists in the dictionary
# If the key exists:
#   → it returns the existing value (23)
#   → it does NOT change the dictionary
# If the key does NOT exist:
#   → it adds the key with the given default value
#   → then returns that value

print(a)
# Prints the value returned by setdefault()
```

## **Nested Dictionaries**

```python
employees = {
    1: {"Name" : "Jhon", "Age" : 23, "Gender" : "Male"}, # 1: is int key of nested dictionary
    2: {"Name" : "Lisa", "Age" : 24, "Gneder" : "Female"}, # 2: is int key of nested dictionary
    3: {"Name" : "Peter", "Age" : 22, "Gender" : "Male"}  # 3: is int key of nested dictionary
}

print (employees[1]["Gender"]) # it will print the Gender of nested dictionary key 1.
print (employees[2]["Name"]) # it will print the Name of nested dictionary key 2.
print (employees [3]["Age"]) # it will print the Age of nested dictionary key 3.
```

## **Problem Solving**

### **1. Write a python program to sort a dictionary by value.**

```python
a = {"a" : 12, "b" : 23, "c" : 6, "d" : 91, "e" : 45}
a = sorted(a.values()) # it will sort dictionary values by numbers and convert it into a list
print (a) 
print (type(a))
```

### **2. Write  a python script to print a dictionary where the keys are numbers between 1 and 15 and the values are square of keys.**

```python
a = {}
# Creates an empty dictionary

for i in range(1, 16):
    # Loop runs from 1 to 15 (16 is not included)
    
    a[i] = i**2
    # Uses i as the key and i squared as the value
    # Example: when i = 3 → key = 3, value = 9

print(a)
# Prints the final dictionary
```

### **3. Write a python program to multiply all the items in a dictionary.**

```python
a = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 5}
# Dictionary where keys are letters and values are numbers

product = 1
# Variable to store the product of all values
# Initialized as 1 because 1 is the identity value for multiplication

for i in a:
    # Loop runs over the dictionary keys: "a", "b", "c", "d", "e"

    product *= a[i]
    # Multiplies current product with the value of key i
    # Example: when i = "c", a["c"] = 3 → product = product * 3

print(product)
# Prints the final product of all values in the dictionary
```

### **4. Write a python program to sort a dictionary by key.**

```python
a = {12: "a", 56: "b", 23: "c", 48: "d", 91: "e"}
# Dictionary where keys are integers and values are strings

a = sorted(a.keys())
# a.keys() → gets all dictionary keys: 12, 56, 23, 48, 91
# sorted() → sorts the keys in ascending order
# Result becomes a LIST, not a dictionary

print(a)
# Prints the sorted list of keys
```

### **5. Convert the following dictionary into JSON format.**

```python
import json
# Import json module to work with JSON conversion

student_data = {"Name": "David", "Age": 13, "Marks": 87}
# student_data is a Python DICTIONARY

print(type(student_data))
# Prints the data type of student_data
# Output: <class 'dict'>

data = json.dumps(student_data)
# json.dumps() converts Python dictionary into a JSON STRING

print(type(data))
# Prints the data type after conversion
# Output: <class 'str'>
```

### **6. Access the value of age from the given JSON data.**

```python
import json
# Import the json module to work with JSON data

student_data = '{"Name": "David", "Age": 13, "Marks": 87}'
# student_data is a STRING
# The string is written in JSON format

data = json.loads(student_data)
# json.loads() converts the JSON string into a Python dictionary
# Now 'data' behaves like a normal Python dict

print(data["Age"])
# Accesses the value of the key "Age" from the dictionary
```

### **7. Pretty print following JSON data.**

```python
import json  
# Imports the json module so we can work with JSON data

student_data = {
    "Name": "David",   # Student name (string)
    "Age": 13,         # Student age (integer)
    "Marks": 87        # Student marks (integer)
}
# This is a Python dictionary storing student information

data = json.dumps(
    student_data,      # The Python dictionary to convert into JSON
    indent=4,          # Makes the output readable with 4 spaces indentation
    separators=(",", "=")  # Uses ',' between items and '=' instead of ':' 
)
# Converts the Python dictionary into a JSON-formatted string

print(data)
# Prints the final JSON-style string on the screen
```

### **8. Sort the following JSON keys and write them into a file.**

```python
import json
# Import json module to work with JSON data

student_data = {
    "Name": "David",   # Student name
    "Age": 13,         # Student age
    "Marks": 87        # Student marks
}
# Python dictionary containing student information

with open("demo.json", 'w') as f:
    # Opens (or creates) a file named demo.json in write mode
    
    data = json.dumps(student_data, indent=4, sort_keys=True)
    # Converts the dictionary into a formatted JSON string
    # indent=4 makes it readable
    # sort_keys=True arranges keys alphabetically
    
    f.write(data)
    # Writes the JSON string into demo.json file
    
    print("The data has been added to the file.")
    # Prints confirmation message
```

### **9. Access the nested key "marks" from the following nested data.**

```python
import json
# Import json module to work with JSON data

student_data = """
{"student": {
    "grade":{
        "name" : "David",
        "marks" : {
            "math" : 87
        }
    }
}}
"""
# student_data is a multiline STRING
# It contains data written in JSON format

data = json.loads(student_data)
# json.loads() converts the JSON string into a Python dictionary
# Now 'data' can be accessed like a normal dictionary

print(data["student"]["grade"]["marks"])
# Accessing nested data step by step:
# data["student"] → gets student dictionary
# ["grade"] → gets grade dictionary
# ["marks"] → gets marks dictionary
```
