## ***Set (`set`)***

- Sets are unordered collection of data. Every element inside the set is unique and mutable.

- Sets are written inside the curly brackets.

- The value inside a set is separated by coma (,).

- Mutable means once created, they can be changed.

```python
names = {"Ali", "Ahmed", "Asad", "Ahsan"}

print (type(names))
```

```python
# empty set

names = set()

print (type(names))
```

```python
names = {"Ali", "Ahmed", "Asad", "Ahsan"}
# Set 'names' contains some unique names

for i in names:
    # Loop through each element in the set 'names'
    # Note: Sets are unordered, so the order of printing may vary
    
    print(i)
    # Prints each element of the set one by one
```

## **Set functions**

```python
# add function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

a.add("Spiderman")  
# add() adds a new element "Spiderman" to the set
# If the element already exists, it will NOT be added again
# The original set 'a' is modified

print(a)
# Prints the updated set 'a' containing the new element
```

```python
# pop function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

a.pop()
# pop() removes and returns an **arbitrary element** from the set
# IMPORTANT: Sets are unordered, so there is no "first element"
# The original set 'a' is modified

print(a)
# Prints the set 'a' after removing one element
```

```python
# remove function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

a.remove("Thor")  
# remove() removes the element "Thor" from the set 'a'
# IMPORTANT: if the element does NOT exist, it will **raise an error**
# The original set 'a' is modified

print(a)
# Prints the set 'a' after removing "Thor"
```

```python
# discard function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

a.discard("Hulk")
# discard() removes the element "Hulk" from the set 'a'
# If the element does NOT exist, no error is raised
# The original set 'a' is modified

print(a)
# Prints the set 'a' after removing "Hulk"
```

```python
# copy function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = a.copy()
# copy() creates a **new set** with the same elements as 'a'
# Changes to 'b' will NOT affect 'a' and vice versa

print(b)
# Prints the new copied set 'b'
```

```python
# isdisjoint function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (Some are also present in set 'a')

print(a.isdisjoint(c))
# isdisjoint() checks whether set 'a' has no elements in common with set 'c'
# Returns True if there are NO common elements, otherwise False
# Original sets are NOT modified
```

```python
# issubset function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (These are also present in set 'a')

print(c.issubset(a))
# issubset() checks if all elements of set 'c' are present in set 'a'
# Returns True if yes, otherwise False
# Original sets are NOT modified
```

```python
# issuperset function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (Note: 'a' is not used in this code)

print(b.issuperset(c))
# issuperset() checks if set 'b' contains all elements of set 'c'
# Returns True if all elements of 'c' are in 'b', otherwise False
# Original sets are NOT modified
```

```python
# Update function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (Note: 'c' is not used in this code)

a.update(b)
# update() adds all elements from set 'b' to set 'a'
# Only unique elements are kept (no duplicates)
# The original set 'a' is modified

print(a)
# Prints the updated set 'a' containing elements from both 'a' and 'b'
```

```python
# Clear function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (Note: 'b' and 'c' are not used in this code)

a.clear()
# clear() removes all elements from the set 'a'
# After this operation, 'a' becomes an empty set

print(a)
# Prints the set 'a', which is now empty
```

```python
# Union function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes
# (Note: 'c' is not used in this code)

print(a.union(b))
# union() returns a new set containing all elements
# from both 'a' and 'b'
# No duplicates are included
# The original sets 'a' and 'b' are NOT modified
```

```python
# Difference function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are also in 'a'

print(a.difference(c))
# difference() returns a new set containing elements
# that are in 'a' but NOT in 'c'
# Does NOT modify the original set 'a'
```

```python
# difference update

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are also in 'a'

a.difference_update(c)
# difference_update() removes elements from 'a' that are also in 'c'
# Only elements not in 'c' remain in 'a'
# This **modifies the original set 'a'**

print(a)
# Prints the updated set 'a' after removing common elements
```

```python
# intersection function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are also in 'a'

x = a.intersection(c)
# intersection() returns a new set containing only elements
# that are present in both 'a' and 'c'
# Does NOT modify the original set 'a'

print(x)
# Prints the new set containing only common elements
```

```python
# intersection update function

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains other superheroes
# (Note: 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are also in 'a'

a.intersection_update(c)
# intersection_update() updates set 'a' directly
# It keeps only elements that are common in both 'a' and 'c'
# Elements not common are removed from 'a'

print(a)
# Prints the updated set 'a' containing only common elements
```

```python
# symmetric difference

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains different superheroes
# (Note: set 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are also in 'a'

x = a.symmetric_difference(c)
# symmetric_difference() returns a new set
# containing elements that are in 'a' OR 'c'
# BUT not in both (removes common elements)
# Does NOT change the original set 'a'

print(x)
# Prints the new set containing only non-common elements
```

```python
# symmetric difference update

a = {"Ironman", "Hulk", "Thor", "Captain America"}
# Set 'a' contains names of some superheroes

b = {"Superman", "Batman", "Wonder-Woman"}
# Set 'b' contains different superheroes
# (Note: set 'b' is not used in this code)

c = {"Hulk", "Thor"}
# Set 'c' contains some superheroes that are common with set 'a'

a.symmetric_difference_update(c)
# symmetric_difference_update() updates set 'a'
# It keeps elements that are in 'a' OR in 'c'
# BUT removes elements that are common in both sets
# Common elements here: "Hulk" and "Thor"

print(a)
# Prints the updated set 'a'
```

# **Problem Solving**

### **1. Write a program to find max and min in a set.**

```python
a = {12, 56, 34, 8, 90, 1, 57}
# Set 'a' contains integer values

maximum = max(a)
# max() finds the largest value present in the set

minimum = min(a)
# min() finds the smallest value present in the set

print("The minimum value in the set is:", minimum)
# Prints the smallest value from the set

print("The maximum value in the set is:", maximum)
# Prints the largest value from the set
```

### **2. Write a program to find difference between two sets.**

```python
a = [1, 5, 6, 8, 2]
# List 'a' contains some integer values

b = [4, 5, 6, 7]
# List 'b' contains some integer values

c = [1, 9, 6, 2, 5]
# List 'c' contains some integer values

print(set(a) & set(b) & set(c))
# set(a), set(b), set(c) convert lists into sets
# '&' is the intersection operator
# It returns only those elements
# which are common in ALL three sets
```

### **3.Write a python program to find difference between two sets.**

```python
a = {1, 5, 6, 8, 2}
# Set 'a' contains some integer elements

b = {4, 5, 6, 7}
# Set 'b' contains some integer elements

print(b.difference(a))
# difference() returns elements that are in set 'b'
# but NOT present in set 'a'
# Common elements (5 and 6) are removed
```

### **4. Write a python program to remove an item from a set if it is present in the set.**

```python
a = {1, 5, 6, 8, 2}
# Set 'a' contains some integer elements

a.discard(12)
# discard() tries to remove the element 12 from the set
# If the element exists, it is removed
# If the element does NOT exist, NO error is raised

print(a)
# Prints the set after discard operation
```

### **5. Write a python program to check if a set is a subset of another set.**

```python
a = {1, 2, 3, 4, 5, 6}
# Set 'a' contains six elements

b = {2, 3, 4}
# Set 'b' contains three elements

print(b.issubset(a))
# issubset() checks whether all elements of set 'b'
# are present in set 'a'
# If yes, it returns True
# If no, it returns False
```
