# **Local and Global Variables**
## **1. Introduction**

Variables store data in a program. Based on where a variable is declared and where it can be accessed, variables are mainly divided into **Local Variables** and **Global Variables**.

---

## **2. Local Variable**

### **Definition**

A **local variable** is declared **inside a function or block** and can be used **only within that function or block**.

**Example (Python)**

```python
def add_numbers():
    a = 10      # local variable
    b = 20      # local variable
    print(a + b)

add_numbers()
# print(a)  ❌ Error: a is not accessible outside the function
```


### **Explanation**

* `a` and `b` are created inside the function
* They can only be used inside `add_numbers()`
* Outside the function, they do not exist

### **Advantages**

* More secure
* Uses less memory
* No naming conflicts

---

## **3. Global Variable**

### **Definition**

A **global variable** is declared **outside all functions** and can be accessed by any function in the program.

**Example (Python)**

```python
x = 50   # global variable

def show_value():
    print(x)

show_value()
```

### **Explanation**

* `x` is declared outside the function
* It can be accessed inside `show_value()`
* It exists for the entire program

---

## **4. Local and Global Variable with Same Name**

```python
x = 100   # global variable

def demo():
    x = 20   # local variable
    print(x)

demo()
print(x)
```

### **Explanation**

* Local `x` is used inside the function
* Global `x` remains unchanged

---

## **5. Using Global Keyword**

### **Why `global` keyword is needed?**

To modify a global variable **inside a function**, the `global` keyword is required.

**Example**

```python
count = 0   # global variable

def increase():
    global count
    count += 1

increase()
print(count)
```


### **Explanation**

* Without `global`, Python creates a new local variable
* `global` tells Python to use the global variable

---

## **6. Comparison Table**

| Feature     | Local Variable  | Global Variable  |
| ----------- | --------------- | ---------------- |
| Declaration | Inside function | Outside function |
| Scope       | Limited         | Whole program    |
| Lifetime    | Temporary       | Entire program   |
| Memory      | Efficient       | Less efficient   |
| Safety      | High            | Low              |

---

## **7. Conclusion**

Local variables are safer and preferred for most tasks. Global variables should be used carefully and only when data needs to be shared across functions.
