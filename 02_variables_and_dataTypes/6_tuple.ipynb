## ***Tuple (`tuple`)***

- Tuples are the collection of ordered and un-mutable data.

- For tuples no brackets are mandatory. By choice one can use parentheses.

- The value inside a Tuple is separated by coma(,).

- Once created, tuples cannot be changed.

- Multiple datatypes can be written inside a tuple.

```python
a = ("apple", "mango", "banana")
print (a)
print (type(a))
```

### **Slicing and Iteration in tuples**

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")
print (a[1:3]) # it will print element at index 1 to 2, index 3 element is excluded
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[:3]) # it will print elements from index 0 to 2, index 3 element is excluded
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[2:]) # it will print elements from index 2 to last index of the tuple
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[::]) # it will print elements from index 0 to last index of the tuple
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[::2]) # it will print elements from index 0 to last index of the tuple, but with a step of 2
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[1::2]) # it will print elements from index 1 to last index of the tuple, but with a step of 2
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[::-1]) # it will print elements from index -1 to first index of the tuple in reverse, but with a step a -1 (reverse the tuple)
```

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

print (a[2::-1]) # it will print elements from index -1 to index 2 of the tuple with a step of -1 in reverse
```

With `for` loop

```python
a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")
# 'a' is a tuple containing names of mobile brands
# Tuples are ordered and immutable (cannot be changed)

for i in a:
    # Loop through each element in the tuple 'a'
    print(i)
    # Prints each mobile brand one by one
```

```python
# Along with range and length in for loop

a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

for i in range (len(a)): # it will access each element in tuple using there index number
    
    print (a[i])
```

With `while` loop

```python
# Along with while loop

a = ("OnePlus", "Vivo", "Redmi", "Samsumg", "Nokia")

i = 0
while i < len (a): # it will access numbers from 0 to len (a)
    print (a[i]) # it will print 
    i += 1
print (i)
```

## **Conversion of Tuples and Tuple Functions**

```python
a = ("onePlus", "Nokia", "Redmi")
print ("Before Conversion: ", type(a))

a = list (a)
print ("After Conversion: ", type(a))

a.append ("Vivo") # it will add "Vivo" at the end of list
a = tuple (a) # it will convert again list into tuple
print (a)
```

```python
a = ("onePlus", "Nokia", "Redmi")

print (a.count("Redmi")) # it will count that how many times there Redmi is in tuple

print (a.index("Nokia")) # it will print index number of nokia

```
