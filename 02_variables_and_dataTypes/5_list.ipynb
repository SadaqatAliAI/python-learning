## _**List (`list`)**_

- Lists are the collection of ordered and mutable data.

- Lists are written inside the squared brackets.

- The value inside a list is separated by coma(,).

- Mutable means once created, they can be changed.

- Multiple datatypes can be written inside a list.

```python
numbers = [1, 2, 3, 4, 5]
# 'numbers' is a list containing integer values

print(numbers)
# Prints the entire list

print(type(numbers))
# type() function tells the data type of 'numbers'
# Since it is a list, the output will be <class 'list'>
```

```python
persons = ["Ali", "Ahmed", "Asad", "Ahsan"]
# 'persons' is a list containing names of people

print(persons)
# Prints the entire list

print(type(persons))
# type() function tells the data type of 'persons'
# Since it is a list, the output will be <class 'list'>
```

```python
fruits = ["apple", "banana", "orange", 23, 34, 45]
# 'fruits' is a list containing both strings (fruit names) and numbers
# Lists in Python can contain elements of different data types

print(fruits)
# Prints the entire list

print(type(fruits))
# type() function tells the data type of 'fruits'
# Since it is a list, the output will be <class 'list'>
```

### **Slicing of list**

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[-3]) # negative slicing (from end) Output: Thor
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[2]) # positive slicing (from start) |  Output: Captain America
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[1:3]) # print from index 1 to 2, index 3 is excluded | Output: Thor, Captain America
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[:2]) # print from index 0 to 1, index 2 is excluded | Output: Ironman, Thor
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[1:]) # print from index 1 to last index | Output: Thor, Captain America, Hulk
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[::3]) # print from index 1 to last index with a step of 3 | Output: Ironman, Hulk
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[-3:-1]) # it will start from end of list and will print from index -3 to -2, index -1 is excluded | Output: Thor, Captain America
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[::-1]) # it will start from the last of the list and print each element (reverse of list)  Output: Hulk, Captain America, Thor, Ironman
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[-1:-4:-1]) # it will start from last of the list and will print from index -1 to -3 with a step of -1, index -4 is excluded | Output: Hulk, Captain America, Thor
```

### **List Iteration**

#### 1. Iteration using for loop.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

for i in a: # will access each element in list one by one
    print (i) # will display each in of list one by one
```

#### 2. Iteration using for loop with range and length function.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

for i in range (1, len(a)): # it will access elements from index 1 to len(a)
    print (a[i]) # it will print elements on index 1 to len(a)
```

#### 3. Iteration using while loop.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

i = 0 # intialize "i" with 0 because index starts from 0
while i < (len(a)): # index is always less than len() becuase index start from 0 while len() starts from 1
    print (a[i])
    i+=1 
```

#### 4. Using shorthand for loop.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]
# This is a list containing superhero names

[print(i) for i in a]
# This is a list comprehension
# It loops through each element in the list 'a'
# For every element 'i', it prints the value
# Output will be each name printed on a new line
```

### **List Functions**

#### 1. To find length of a list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]
# This is a list containing 4 elements (superhero names)

print(len(a))
# len() function is used to find the total number of elements in the list
# Here, the list has 4 items, so the output will be 4
```

#### 2. To count an occurence of a particular element.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]  
# This is a list that contains names of superheroes

print(a.count("Hulk"))  
# count() function checks how many times "Hulk" appears in the list
# Since "Hulk" is present only one time, the output will be 1
```

#### 3. To add to the list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.append("Spiderman") # it will add Spiderman at the end of the list
print (a)
```

#### 4. To add to a specific location.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.insert(1, "Vision") # it will add Vision at index 1
print (a)
```

#### 5. To remove from a list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.remove("Hulk") # it will remove Hulk from list
print (a)
```

#### 6. To remove from a certain location.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.pop (1) # it will remove element from index 1.
print (a)
```

#### 7. To create a copy of a list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]
# Original list

b = a.copy()
# copy() creates a NEW list with the same elements

print(b)
```

#### 8. To find/access the index number of an element.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a.index("Hulk")) # it will give the index number of Hulk

```

#### 9. To extend the list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

c = ["Vision", "Spiderman"]

a.extend(c) # it will add list c into list a at last of list a

print (a)
```

#### 10. To reverse the list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

print (a[::-1]) # it will reverse the original list
```

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.reverse() # it will reverse the original list
print (a)
```

#### 11. To sort the list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]

a.sort() # it will sort the elements of list wrt alphbets
print (a)
```

```python
a = [4,5,3,2,7,8,5,2]

a.sort() # it will sort list wrt to numbers
print (a)
```

#### 12. To clear all the data from list.

```python
a = ["Ironman", "Thor", "Captain America", "Hulk"]
# List 'a' contains 4 elements

a.clear()
# clear() removes ALL elements from the list
# The list itself still exists, but it becomes empty

print(a)
# Prints the empty list
```

### **List Comprehension**

```python
l1 = [30, 40, 50, 60]

l2 = []
for i in l1: # it will access each element of l1 one by one
    l2.append(i) # it will append/add each element of l1 accessed using i at the end
print (l2)
```

```python
l1 = [30, 40, 50, 60]

l2 = []
for i in l1: # it will access each element of l1 one by one
    if (i > 45):
        l2.append(i) # it will append/add those  elements of l1 which are greater than 45 accessed using i at the end
print (l2)
```

```python
# Short Hand

l1 = [30, 40, 50, 60]

l2 = [i for i in l1 ] # it will add each element of l1 to l2

print (l2)
```

```python
# Short Hand

l1 = [30, 40, 50, 60]

l2 = [i for i in l1 if i > 35] # it will add only those elements of l1 in l2 which are greater than 35

print (l2)
```

### **5. Write a python program to get the largest number from the list.**

```python
# Solution 1 

b = [13, 7, 12, 10]

largest = b[0] # it will let index 0 element largest one
i = 0
for i in range(0, len(b)): # it will access each element in the list by there index number
    if (b[i]> b[0] ): # it will check if any of the index in the loop is greater than considered b[0] largest, then it will become largest number
        largest = b[i] # assigning value
print ("Largest number: ", largest)
```

```python
# Solution 2

b = [13, 7, 12, 10]
b.sort() 
print ("Largest number: ", b[-1])
```

### **6. Write a python program to find the smallest number from the list.**

```python
# Solution 1
b = [13, 7, 12, 10]

i = 0 # it will intialize index with 0
smallest = b[0] # let index 0 element is smallest
for i in range (0, len(b)): # it will access each element in the list by there index number
    if (b[i] < b[0]): # it will check if any of the index in the loop is smaller than considered b[0] smallest, then it will become smallest number
        smallest = b[i] # assigning value
print ("Smallest number: ", smallest)
```

```python
# Solution 2

b = [13, 7, 12, 10]
b.sort()
print ("Smallest number: ", b[0])
```
