

## ✅ 1. What is String Formatting?
---
📌 **String formatting** means inserting variables or values inside strings in a neat and controlled way.

🔧 It helps you **combine text and data** (like numbers, variables) in a readable format.

---

## ✅ 2. 3 Common Ways of String Formatting in Python

### ✅ Method 1: Using **f-strings** (✅ Recommended – Easy and Fast 🏎️)

```python
name = "Sadaqat"
age = 25
print(f"My name is {name} and I am {age} years old.")
```

📌 Output:

```
My name is Sadaqat and I am 25 years old.
```

📘 **Inside `{}`** you can use variables or even expressions like `{age+5}`.

---

### ✅ Method 2: Using `.format()` Method (Older but still useful)

```python
name = "Sadaqat"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
```

📌 Output:

```
My name is Sadaqat and I am 25 years old.
```

You can also specify **position**:

```python
print("My name is {1} and I am {0} years old.".format(age, name))
```

Output:

```
My name is Sadaqat and I am 25 years old.
```

---

### ✅ Method 3: Using **% Operator** (Very old style – not recommended for new code)

```python
name = "Sadaqat"
age = 25
print("My name is %s and I am %d years old." % (name, age))
```

📌 `%s` for string, `%d` for integer, `%f` for float

Output:

```
My name is Sadaqat and I am 25 years old.
```

---

## ✅ 3. Comparison Table

| Method         | Style           | Easy to Use | Recommended  |
| -------------- | --------------- | ----------- | ------------ |
| f-strings      | `f"{var}"`      | ✅✅✅         | ✅✅✅ (Best)   |
| `.format()`    | `"{}".format()` | ✅✅          | ✅            |
| `%` formatting | `"Hello %s"`    | ❌ Old       | ❌ No (Avoid) |

## ✅ 4. More Powerful f-string Tricks

🧮 **Math inside f-string**:

```python
a = 5
b = 3
print(f"Sum is {a + b}")
# Output: Sum is 8
```

📏 **Formatting decimals**:

```python
pi = 3.14159
print(f"Pi is {pi:.2f}")
# Output: Pi is 3.14
```

🕒 **Using datetime**:

```python
import datetime
today = datetime.date.today()
print(f"Today is {today:%A, %d %B %Y}")
# Output: Today is Wednesday, 26 June 2025
```
