### **Use of `sep=`**

- To change how printed values are separated.
- Useful for dates, file paths, formatting strings, and joining values.
- Makes code cleaner when you don't want to use + or manual concatenation

> **Example 1:**

```python
# Without sep=

print ("A", "B", "C")

# Output:
# A B C
```

```python
# With sep=

print ("A","B","C", sep=", ")

# Output:
# A, B, C
```

> **Example 2:**

```python
# Without sep=

print ("Python", "is", "awesome")

# Output:
# Python is awesome
```

```python
# With sep=

print ("Python", "is", "awesome", sep="🔥")

# Output:
# Python🔥is🔥awesome
```

```python
# Without sep=

print ("30","07","2025")

# Output:
# 30 07 2025
```

```python
# With sep=

print ("30", "07", "2025", sep="/")

# Output:
# 30/07/2025
```

### **Use of `end=`**

- To control line endings in output.
- To print multiple things on the same line.
- To format output in a custom and flexible way 



> **Exampel 01:**

```python
# Without end= parameter

print ("Hello")
print ("World")

# Output:
# Hello
# World 
# (on multiple lines)

```

```python
# With end= parameter

print ("Hello", end=" ")
print ("World")

# Output:
# Hello World 
# (on single line)
```

> **Exampel 02:**

```python
# Without end= parameter

print ("A")
print ("B")
print ("C")

# Output:
# A
# B
# C
# (in multiple lines)
```

```python
# With end= parameter

print ("A", end=", ")
print ("B", end=", ")
print ("C")

# Output:
# A, B, C
# (in single lines with ,)
```

> **Example 03:**

```python
# Without end= parameter

print ("Same line")
print ("continue here")

# Output:
# Same line
# continue here
```

```python
# With end parameter

print ("Same line", end=" ")
print ("continue here!")

# Output:
# Same line continue here!
```
